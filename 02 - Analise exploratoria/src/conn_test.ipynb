```python
import networkx as nx
import matplotlib.pyplot as plt

# define the data
schools = ['School A', 'School B', 'School C', 'School D', 'School E',
           'School F', 'School G', 'School H', 'School I', 'School J']
similarities = [['School B', 'School D', 'School E'],
                ['School A', 'School E', 'School H'],
                ['School D', 'School F', 'School J'],
                ['School A', 'School C', 'School J'],
                ['School A', 'School B', 'School G'],
                ['School C', 'School E', 'School H'],
                ['School F', 'School G', 'School I'],
                ['School B', 'School F', 'School I'],
                ['School C', 'School G', 'School J'],
                ['School D', 'School E', 'School H']]

# create a directed graph
G = nx.DiGraph()

# add nodes to the graph
for school in schools:
    G.add_node(school)

# add edges to the graph
for i, school in enumerate(schools):
    for similar_school in similarities[i]:
        G.add_edge(school, similar_school)

# define edge width based on edge weight
edge_width = [(1 + G.degree(edge[0]))*0.5 for edge in G.edges()]

# draw the graph
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_edges(G, pos, width=edge_width)
nx.draw_networkx_labels(G, pos)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# define the data
schools = ['School A', 'School B', 'School C', 'School D', 'School E',
           'School F', 'School G', 'School H', 'School I', 'School J']
similarities = [['School B', 'School D', 'School E'],
                ['School A', 'School E', 'School H'],
                ['School D', 'School F', 'School J'],
                ['School A', 'School C', 'School J'],
                ['School A', 'School B', 'School G'],
                ['School C', 'School E', 'School H'],
                ['School F', 'School G', 'School I'],
                ['School B', 'School F', 'School I'],
                ['School C', 'School G', 'School J'],
                ['School D', 'School E', 'School H']]

# create a directed graph
G = nx.DiGraph()

# add nodes to the graph
for school in schools:
    G.add_node(school)

# add edges to the graph
for i, school in enumerate(schools):
    for similar_school in similarities[i]:
        G.add_edge(school, similar_school)

# define colors for each edge
edge_colors = []
for i, school in enumerate(schools):
    for j, similar_school in enumerate(similarities[i]):
        if school == 'School A':
            if similar_school == 'School B':
                edge_colors.append('red')
            elif similar_school == 'School D':
                edge_colors.append('blue')
            elif similar_school == 'School E':
                edge_colors.append('green')
            else:
                edge_colors.append('gray')
        else:
            edge_colors.append('gray')

# draw the graph
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_edges(G, pos, edge_color=edge_colors)
nx.draw_networkx_labels(G, pos)
```
